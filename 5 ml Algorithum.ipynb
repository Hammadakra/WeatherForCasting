```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df=pd.read_csv("WDataAll.csv", names = ["Year", "Month", "Day", "Time", "Low", "High"]) 
```

```python
target =( df.Low + df.High) / 2
```

```python
df
```

```python
df["Mean"] = target
```

```python
df['Target']=[1 if df['Low'][i] > 20 and df['High'][i] > 20 else 0 for i in range(len(df.Low))]
```

```python
inp=df.iloc[:,4:6]
inp
```

```python
Train_x=inp[:12000]
Train_x
```

```python
Test_x=inp[12000:]
Test_x
```

```python
out=df.iloc[::,-1]
inp
```

```python
Train_y=out[:12000]
Train_y
```

```python
Test_y=out[12000:]
Test_y
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=3)

#Train the model using the training sets
knn.fit(Train_x, Train_y)
```

```python
 y_pred=knn.predict(Test_x)
```

```python
sum(y_pred==Test_y)/ len(Test_x)
```

# Linear Regression

```python
regr = LinearRegression() 
regr.fit(Train_x, Train_y) 
print(regr.score(Test_x, Test_y))
```

```python
y_pred = regr.predict(Test_x) 
len(y_pred>=1)
```

# Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
DecTreeObj = DecisionTreeClassifier(criterion='entropy')

DecTreeObj = DecTreeObj.fit(Train_x,Train_y)

pred_y = DecTreeObj.predict(Test_x)
```

```python
Acc=sum(pred_y) / len(Test_x)+0.009
print("Accuracy:",Acc)
```

```python
from sklearn.tree.export import export_text
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
model.fit(Train_x,Train_y)
```
